```python
# latestkey 크롤링 데이터 중복값 제거 등 전처리
csv_test = pd.read_csv('C:/project/CODE/keyword/data/book/naver_top_500/'+ yesterday +'_naver_book_top_500.csv', header = None, index_col=0)
csv_test.columns = ['rank', 'keyword', 'year', 'month', 'day', 'ymd']
csv_test = csv_test.drop_duplicates(['rank', 'keyword', 'year', 'month', 'day', 'ymd'], keep='first') # 크롤링 과정에서 입력된 중복 행 제거 처리
csv_test['ymd'] = csv_test['ymd'].str.replace('/', '-')
csv_test = csv_test.reset_index(drop=True)
# csv_test = csv_test[csv_test['ymd'] == yesterday] # 최신 일자 데이터 정제 확인
csv_test
```

```python
# 숫자나 알파벳으로 시작하는 상품명키워드들 삭제하기

# df[df.keyword.str.startswith('1')]
df_drop = pd.DataFrame(columns=['rank', 'keyword', 'year', 'month', 'day', 'ymd'])

# 삭제조건 리스트 생성
sample_string = "1234567890abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*()_{}[]<>?"
string_list = []

for word in sample_string:
    string_list.append(word)

for i in range(len(sample_string)):
    df1 = csv_test[csv_test.keyword.str.startswith(string_list[i])]
    df_drop = pd.concat([df_drop, df1])
df_drop

# 삭제할 키워드 인덱스 정보가 있는 변수 생성

idx = df_drop.index
idx

df_clone = csv_test
df_clone = df_clone.drop(idx)
df_clone = df_clone.astype(object)
df_clone
```

```python
# select문 만들기
def select_command(table) :
    command = 'select * from ' + table
    return command

# insert문 만들기
def insert_command(table) :
    command = 'insert into ' + table + ' (rank, keyword, year, month, day, ymd) values (%s, %s, %s, %s, %s, %s);'
    return command
```

```python
# postgresql 접속
conn_string = "host='localhost' dbname='postgres' user='postgres' password='admin'"
conn = psycopg2.connect(conn_string)
cur = conn.cursor()

table = input("테이블명 입력: ") # keyword4 입력

# table에 데이터 넣고 정렬하기
for i in range(len(df_clone)) :
    cur.execute(insert_command(table), (df_clone.iloc[i][0], df_clone.iloc[i][1], df_clone.iloc[i][2], df_clone.iloc[i][3], df_clone.iloc[i][4], df_clone.iloc[i][5]))
    conn.commit()
```
